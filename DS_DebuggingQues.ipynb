```python
#corrected code
import pandas as pd  

data = {'Name': ['Alice', 'Bob', 'Charlie'], 
        'Age': [25, 30, 'Thirty-Five'],  
        'Gender': ['F', 'M', 'M']}  

df = pd.DataFrame(data)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

average_age = df['Age'].mean()

print(f"Average Age: {average_age}")
```
